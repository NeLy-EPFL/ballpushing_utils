```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset:

UMAP = pd.read_feather("/home/matthias/ballpushing_utils/tests/integration/outputs/umap_TNT_1.feather")
```

```python
# Plot UMAP:
sns.set(style="white", palette="muted")
sns.set_context("notebook", font_scale=1.5)
sns.set_style("whitegrid")
sns.set_palette("deep")
sns.set(rc={"figure.figsize": (10, 8)})

sns.scatterplot(
    data=UMAP,
    x="UMAP1",
    y="UMAP2",
    hue="cluster",
    palette=sns.color_palette("deep", n_colors=UMAP["cluster"].nunique()),
    alpha=0.8,
    edgecolor="w",
    linewidth=0.5,
)
sns.despine()

plt.show()
```

```python
Transposed = pd.read_feather("/mnt/upramdya_data/MD/Ballpushing_TNTScreen/Datasets/250418_summary_TNT_screen_Data/transposed/pooled_transposed.feather")
# Cleanup data
def cleanup_data(df):
    # Remove the non-TNT data from the Disp_Data
    df = df[~df["Genotype"].isin(["M6", "M7", "PR", "CS", "TNTxM6", "TNTxCS"])]
    df = df[df["Brain region"] != "None"]
    return df

# Cleanup the data
Transposed = cleanup_data(Transposed)
# Check the individual Genotypes
print(Transposed["Genotype"].unique())

# Save the cleaned data
Transposed.to_feather("/mnt/upramdya_data/MD/Ballpushing_TNTScreen/Datasets/250418_summary_TNT_screen_Data/transposed/pooled_transposed_cleaned.feather")
```
